```python
import pandas as pd
import numpy as np
from imdb import Cinemagoer
import matplotlib.pyplot as plt
import datetime
from collections import Counter

from cleaning_funcs import load_data

ia = Cinemagoer()

%matplotlib inline

#IMDb Docs: https://imdbpy.readthedocs.io/en/latest/index.html
```

```python
data = load_data('movie_list_current.csv')
data = data[['Month', 'Pick', 'IMDb ID', 'Movie', 'Year', 'IMDb Score', 'Steven', 'Jon', 'Farzad', 'Average']]
data = data[data.Steven.notnull()]
print("Number of Movies:", data.shape[0])
# data
```

```python
data = data.rename(columns=str.lower)
data = data.rename(columns={"imdb id": "imdb_id"})
data = data.dropna(subset=['steven', 'jon', 'farzad', 'average'])
data.head()
```

```python
# subset = data[0:3]
subset = data

title_list = []
year_list = []
rating_list = []
runtimes_list = []
cum_ww_gross_list = []
budget_list = []

for index, row in subset.iterrows():
    movie = ia.get_movie(row['imdb_id'])
    
    title_list.append(movie.get('title'))
    year_list.append(movie.get('year'))
    rating_list.append(movie.get('rating'))
    runtimes_list.append(float(movie.get('runtimes')[0]))
    cum_ww_gross_list.append(movie.get('box office')['Cumulative Worldwide Gross'])
    budget_list.append(movie.get('box office')['Budget'])
    # movie_data['imdbid'] = movie.get('imdbid')
    # movie_data['directors'] = movie.get('directors') #need to parse this before we can use it
    # movie_data['genres'] = movie.get('genres')
    # movie_data['runtimes'] = movie.get('runtimes')
    # movie_data['budget'] = movie.get('budget')
    # movie_data['gross'] = movie.get('gross')
    # movie_data['plot'] = movie.get('plot')
    # movie_data['synopsis'] = movie.get('synopsis')

    # temp.append([title, year, rating], ignore_index=True)

subset['title'] = title_list
subset['year'] = year_list
subset['rating'] = rating_list
subset['runtimes'] = runtimes_list
subset['cum_worldwide_gross'] = cum_ww_gross_list
subset['budget'] = budget_list
```
